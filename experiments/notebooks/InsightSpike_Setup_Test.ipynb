## 4. „Éë„Éï„Ç©„Éº„Éû„É≥„Çπ„ÉÜ„Çπ„Éà ‚ö°

GPUÂà©Áî®Áä∂Ê≥Å„ÄÅ„É°„É¢„É™‰ΩøÁî®Èáè„ÄÅÂá¶ÁêÜÈÄüÂ∫¶„Å™„Å©„ÅÆ„Éë„Éï„Ç©„Éº„Éû„É≥„ÇπÊåáÊ®ô„Çí„ÉÅ„Çß„ÉÉ„ÇØ„Åó„Åæ„Åô„ÄÇ

```python
# „Éë„Éï„Ç©„Éº„Éû„É≥„ÇπÁõ£Ë¶ñ„Å®„ÉÜ„Çπ„Éà
import time
import psutil
import gc

print("‚ö° Performance and Resource Testing...")

# „É°„É¢„É™‰ΩøÁî®Èáè„ÇíÂèñÂæó
def get_memory_usage():
    """ÁèæÂú®„ÅÆ„É°„É¢„É™‰ΩøÁî®Èáè„ÇíÂèñÂæó"""
    process = psutil.Process()
    return process.memory_info().rss / 1024 / 1024  # MB

# GPU „É°„É¢„É™‰ΩøÁî®Èáè„ÇíÂèñÂæóÔºàÂà©Áî®ÂèØËÉΩ„Å™Â†¥ÂêàÔºâ
def get_gpu_memory():
    """GPU „É°„É¢„É™‰ΩøÁî®Èáè„ÇíÂèñÂæó"""
    if torch.cuda.is_available():
        return torch.cuda.memory_allocated() / 1024 / 1024  # MB
    return 0

print("üìä System Resource Status:")
print(f"üêè RAM Usage: {get_memory_usage():.1f} MB")
print(f"üéÆ GPU Memory: {get_gpu_memory():.1f} MB")

# Ë§áÊï∞„ÅÆË≥™Âïè„ÅßÂá¶ÁêÜÊôÇÈñì„Çí„ÉÜ„Çπ„Éà
if 'agent' in locals():
    print("\n‚è±Ô∏è  Processing Speed Test:")
    
    speed_test_questions = [
        "Hello",
        "What is machine learning?",
        "Explain the difference between AI and ML in detail.",
    ]
    
    for i, question in enumerate(speed_test_questions, 1):
        gc.collect()  # „Ç¨„Éô„Éº„Ç∏„Ç≥„É¨„ÇØ„Ç∑„Éß„É≥
        if torch.cuda.is_available():
            torch.cuda.empty_cache()  # GPU „Ç≠„É£„ÉÉ„Ç∑„É•„ÇØ„É™„Ç¢
            
        start_time = time.time()
        start_memory = get_memory_usage()
        start_gpu_memory = get_gpu_memory()
        
        try:
            result = agent.process_question(question)
            
            end_time = time.time()
            end_memory = get_memory_usage()
            end_gpu_memory = get_gpu_memory()
            
            processing_time = end_time - start_time
            memory_delta = end_memory - start_memory
            gpu_memory_delta = end_gpu_memory - start_gpu_memory
            
            print(f"\nüìù Speed Test {i}: '{question[:30]}...'")
            print(f"  ‚è±Ô∏è  Time: {processing_time:.2f}s")
            print(f"  üêè Memory Œî: {memory_delta:+.1f} MB")
            print(f"  üéÆ GPU Memory Œî: {gpu_memory_delta:+.1f} MB")
            print(f"  ‚úÖ Success: {result.get('success', False)}")
            
        except Exception as e:
            print(f"  ‚ùå Speed Test {i} failed: {e}")

print("\nüéØ Performance test completed!")
```
